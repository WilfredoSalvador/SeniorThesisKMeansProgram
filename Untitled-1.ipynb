Elbow Method of finding clusters

```python
inertias = []

k_range = range(2, 15)

for k in k_range:
    km_model = KMeans(n_clusters=k, n_init="auto")
    km_model.fit(new_df)
    intertias.append([k, km_model.inertia_])
```

```python
inertias = np.array(inertias)
plt.plot(inertias[:,0], inertias[:,1], 'x-')
plt.show()
```
